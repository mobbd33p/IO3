```python
# Crear la instancia map en m
m = folium.Map(location=[-17.208164789600705, -70.95014545457539],
    zoom_start=19, control_scale=True)

# Añadir Marcador
# Para ayuda en iconos: help(folium.Icon) 
folium.Marker(
    location=[-17.193718443023055, -70.9338005090304],
    popup='Catedral de Moquegua',
    icon=folium.Icon(color='green', icon='ok-sign'),
).add_to(m)

#Ver Mapa
m
```

```python
 #Para abrir excel
 df = pd.read_excel('MapsLugares.xlsx')
 #Para abrir CSV
 #df = pd.read_csv('MapsLugares.csv')
df.head()
```

```python
m = folium.Map([-17.193038956670495, -70.9318361402748],  zoom_start=14)

for index, row in df.iterrows():
    folium.Marker([row['Latitud'], row['Longitud']], 
                  popup=row['Lugar'],
                  icon=folium.Icon(icon='Icono')
                 ).add_to(m)
m
```
